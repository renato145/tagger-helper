```python
# count json files
import os
folder = 'train'
folders = [os.path.join(folder, f) for f in os.listdir(folder) if os.path.isdir(os.path.join(folder, f))]
for f in folders:
    n_json = len([x for x in os.listdir(f) if x.find('.json') > -1])
    n_imgs = len([x for x in os.listdir(f) if x.find('.jpg') > -1])
    n_npy = len([x for x in os.listdir(f) if x.find('.npy') > -1])
    print('%-11s: %d/%d - %d' % (f, n_json, n_imgs, n_npy))
```
